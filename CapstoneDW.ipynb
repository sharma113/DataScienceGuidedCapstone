```python
#Create Look up Tables
id_admission_type = id.iloc[0:9,:]
```

```python
#Create Look up Tables
discharge_disposition_id = id.iloc[10:41,:]
```

```python
#Create Look up Tables
admission_source_id = id.iloc[42:68,:]
```

```python
#Rename Columns
discharge_disposition_id = discharge_disposition_id.rename(columns={'ID mappings for the diabetes dataset': 'discharge_disposition_id', ' ? - always means a missing value': 'discharge_disposition_description'})
```

```python
#reset index
discharge_disposition_id = discharge_disposition_id.reset_index(drop=True)
#Drop First row
discharge_disposition_id = discharge_disposition_id.drop(0)
```

```python
#Create look up table.  Reset Index, Drop first Row.
admission_source_id = admission_source_id.rename(columns={'ID mappings for the diabetes dataset': 'admission_source_id', ' ? - always means a missing value': 'admission_source_description'})
admission_source_id = admission_source_id.reset_index(drop=True)
admission_source_id = admission_source_id.drop(0)
```

```python
#print(discharge_disposition_id.columns)
```

```python
#Create look up table.  Reset Index, Drop first Row.
admission_type_id =  id_admission_type
admission_type_id = admission_type_id.rename(columns={'ID mappings for the diabetes dataset': 'admission_type_id', ' ? - always means a missing value': 'admission_type_description'})
admission_type_id = admission_type_id.reset_index(drop=True)
admission_type_id = admission_type_id.drop(0)
```

```python
#Change Datatype to Int
discharge_disposition_id['discharge_disposition_id'] = discharge_disposition_id['discharge_disposition_id'].astype(int)
admission_type_id['admission_type_id'] = admission_type_id['admission_type_id'].astype(int)
admission_source_id['admission_source_id'] = admission_source_id['admission_source_id'].astype(int)
```

```python
#Change Description Values
discharge_disposition_id['discharge_disposition_description'] = discharge_disposition_id['discharge_disposition_description'].fillna(0)
discharge_disposition_id['discharge_disposition_description'] = np.where(discharge_disposition_id['discharge_disposition_description'] == 'Not Mapped', 'Not Available', discharge_disposition_id['discharge_disposition_description'])
discharge_disposition_id['discharge_disposition_description'] = np.where(discharge_disposition_id['discharge_disposition_description'] == 0, 'Not Available', discharge_disposition_id['discharge_disposition_description'])
discharge_disposition_id['discharge_disposition_description'] = np.where(discharge_disposition_id['discharge_disposition_description'] == 'Unknown/Invalid', 'Not Available', discharge_disposition_id['discharge_disposition_description'])
 
```

```python
#Change Unknown Description Values
admission_type_id['admission_type_description'] = admission_type_id['admission_type_description'].fillna(0)
admission_type_id['admission_type_description'] = np.where(admission_type_id['admission_type_description'] == 'Not Mapped', 'Not Available', admission_type_id['admission_type_description'])
admission_type_id['admission_type_description'] = np.where(admission_type_id['admission_type_description'] == 0, 'Not Available', admission_type_id['admission_type_description'])
print(admission_type_id)
```

```python
#Change Unknown Description Values
admission_source_id['admission_source_description'] = admission_source_id['admission_source_description'].fillna(0)
admission_source_id['admission_source_description'] = np.where(admission_source_id['admission_source_description'] == 'Not Mapped', 'Not Available', admission_source_id['admission_source_description'])
admission_source_id['admission_source_description'] = np.where(admission_source_id['admission_source_description'] == 'Unknown/Invalid', 'Not Available', admission_source_id['admission_source_description'])
admission_source_id['admission_source_description'] = np.where(admission_source_id['admission_source_description'] == 0, 'Not Available', admission_source_id['admission_source_description'])

print(admission_source_id)
```

```python
#Perform Merge to bring look up tables into the main DataFrame
merged_df = pd.merge(df, discharge_disposition_id, on='discharge_disposition_id',how='inner')
merged_df = pd.merge(merged_df, admission_source_id, on='admission_source_id',how='inner')
merged_df = pd.merge(merged_df, admission_type_id, on='admission_type_id',how='inner')
```
